```python
import matplotlib.pyplot as plt
import numpy as np

# Define los parámetros del problema
radio = 2
alpha = 0.03  # por ejemplo
K = 4  # por ejemplo

# Crea matrices para contener los ángulos de inicio y final de cada segmento
angulos_inicio = np.zeros(2*K)
angulos_final = np.zeros(2*K)

# Rellena las matrices
for i in range(K):
    angulos_inicio[2*i] = i*2*np.pi/K
    angulos_inicio[2*i+1] = angulos_inicio[2*i] + alpha*2*np.pi

    angulos_final[2*i] = angulos_inicio[2*i+1]
    angulos_final[2*i+1] = (i+1)*2*np.pi/K

# Define los colores para los segmentos de electrodos y no electrodos
colores = ['b' if i%2==0 else 'r' for i in range(2*K)]

# Traza los segmentos del círculo
fig, ax = plt.subplots(figsize=(6, 6))

for i in range(2*K):
    angulos = np.linspace(angulos_inicio[i], angulos_final[i], 100)
    x = radio * np.cos(angulos)
    y = radio * np.sin(angulos)
    ax.plot(x, y, color=colores[i])
    
    # Agrega una etiqueta en medio de cada segmento de electrodo
    if i%2 == 0:  # Solo etiqueta segmentos de electrodo
        angulo_medio = 0.5 * (angulos_inicio[i] + angulos_final[i])
        ax.text(radio * np.cos(angulo_medio), radio * np.sin(angulo_medio), f'Electrodo {i//2+1}', 
                horizontalalignment='center', verticalalignment='center')

plt.axis('equal')  # Aspecto igual
plt.show()
```
